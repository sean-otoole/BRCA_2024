```python
## first I want to import the bulk signatures and make sure that it works properly

import pyreadr
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import to_hex
from matplotlib.colors import to_rgba 


bulk_sigs = pyreadr.read_r('/tmp/work/Visium/BRCA_2024/gene_signatures/all_signature_scores.Rds') # also works for RData
bulk_sigs = bulk_sigs[None] # extract the pandas data frame 

# Scale the data
# scaler = StandardScaler()
# scaled_data = scaler.fit_transform(bulk_sigs)

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(bulk_sigs)

# Create a DataFrame with PCA results
rownames = bulk_sigs.index.tolist()  # Get row names
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'], index=rownames)

# Perform k-means clustering
kmeans = KMeans(n_clusters=2, random_state=123)
pca_df['cluster'] = kmeans.fit_predict(pca_df[['PC1', 'PC2']])
```

```python
palette = sns.color_palette("hsv", 2)

# Create scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(
    x='PC1', y='PC2', hue='cluster',
    palette=palette, data=pca_df,
    s=100, edgecolor='w', alpha=0.7
)

# Add labels to the plot
for i in range(pca_df.shape[0]):
    scatter_plot.text(
        pca_df.PC1.iloc[i], pca_df.PC2.iloc[i],
        rownames[i],
        fontsize=9, ha='right'
    )

plt.title('PCA Plot with K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(False)
plt.show()
```

```python
bulk_sigs['cluster'] = pca_df['cluster']
bulk_sigs_sorted = bulk_sigs.sort_values(by='cluster')

cluster_df = bulk_sigs_sorted['cluster']
cluster_df = pd.DataFrame(cluster_df)

bulk_sigs_sorted = bulk_sigs_sorted.drop(columns='cluster')

# Map cluster numbers to colors
unique_clusters = sorted(cluster_df['cluster'].unique())

cluster_colors = {1: to_rgba('green'), 2: to_rgba('blue')}


# Set the figure size

plt.figure(figsize=(16, 10))  # Increased width for better visibility of annotation


# Plot heatmap with annotations
sns.heatmap(
    bulk_sigs_sorted,
    annot=False,                  # Show values in cells
    cmap='coolwarm',              # Color map
    xticklabels=True,            # Show all x-tick labels
    yticklabels=True,            # Show all y-tick labels
    cbar_kws={'label': 'Scale'}  # Color bar label
)

# Add annotation bar on the side for clusters
ax = plt.gca()
for i, cluster in enumerate(cluster_df['cluster']):
    ax.add_patch(plt.Rectangle(
        (bulk_sigs_sorted.shape[1], i), width=1, height=1, color = cluster_colors[cluster+1]
    ))

# Adjust plot limits to make space for the annotation bar
plt.xlim(-1, bulk_sigs_sorted.shape[1] + 1)

plt.title('Heatmap of Ordered DataFrame with Cluster Annotations')
plt.xlabel('Features')
plt.ylabel('Samples')

# Ensure only unique clusters from the data are in the legend
unique_clusters = sorted(cluster_df['cluster'].unique())

# Add a legend for clusters
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=cluster_colors[cluster+1], markersize=10) for cluster in unique_clusters]
plt.legend(handles, [f'Cluster {cluster}' for cluster in unique_clusters], title='Clusters', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# All possible categories
cell_type_categories = ['Epithelial cells', 'Endothelial cells', 'Fibroblasts', 'B cells', 'Plasma cells', 'Macrophages','pDC']

# Initialize an empty DataFrame to store the results
combined_df = pd.DataFrame(index=cell_type_categories)

# Loop over each dataset
for sample in sample_names:
    # Create the frequency table and normalize it to relative frequencies
    current_data_set = visium_samples[sample]
    current_counts = current_data_set.obs['majority_voting']
    frequency_table = current_counts.value_counts().reindex(cell_type_categories, fill_value=0)
    relative_frequencies = frequency_table / frequency_table.sum()
    combined_df[sample] = relative_frequencies

print(combined_df)

simple_cateogories = {'epi':combined_df.iloc[0:1,:].sum(axis = 0),
            'all_others':combined_df.iloc[1:,:].sum(axis = 0)}
simple_df = pd.DataFrame(data=simple_cateogories)
simple_df = simple_df.transpose()

print(simple_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib.colors import ListedColormap

# Create the primary heatmap using clustermap
plt.figure(figsize=(10, 6))  # Increase figure width to accommodate both heatmaps
clusterMapPlot = sb.clustermap(
    signature_scores_df_z_by_cluster,
    annot=False,
    cmap='coolwarm',
    fmt='.2f',
    col_cluster=True,
    row_cluster=False,
    yticklabels=True  # Show row labels in the primary heatmap
)
plt.rcParams["axes.grid"] = False

# df2 contains the cluster information (0 or 1)
df2 = subset_df[['cluster']]

# Create a custom colormap for the clusters
cluster_colors = ListedColormap(['blue', 'green'])

# Place the second heatmap farther away from the first one
heatmap_ax = clusterMapPlot.ax_heatmap  # The axis of the main heatmap
x, y, width, height = heatmap_ax.get_position().bounds

# Add a new axis for the second heatmap with increased space
ax2 = clusterMapPlot.fig.add_axes([x + width + 0.1, y, 0.05, height])  # Increase spacing here

# Plot the second heatmap with the custom colormap
sb.heatmap(
    df2,
    cmap=cluster_colors,
    cbar=True,
    ax=ax2,
    annot=False,
    yticklabels=False  # Hide row labels in the second heatmap
)

# Adjust colorbar position
colorbar = ax2.collections[0].colorbar
colorbar.ax.set_ylabel('Cluster', rotation=270, labelpad=15)

# Adjust the title to avoid overlap
clusterMapPlot.fig.suptitle('Heatmap with Adjacent Cluster Heatmap', y=1.02)

# Show the plot
plt.show()
```
